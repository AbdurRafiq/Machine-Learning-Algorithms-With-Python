# Introduction to Scikit -Learn (Sklearn)



```python
# 1 get the data ready

import pandas as pd
import numpy as np
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
#Create X( features matrix)

X = heart_disease.drop("target",axis=1)

# Create y (labels)

y = heart_disease["target"]
```

```python
# 2. choose the right modele and hyperparameters

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()

# we'll keep default hyperparameters

clf.get_params()
```

```python
# 3. fit the modelto the training data
from sklearn.model_selection import train_test_split

X_train,X_test,y_train, y_test = train_test_split(X,y,test_size = 0.25)
```

```python
X_train
```

```python
clf.fit(X_train,y_train);
```

```python
#make A prediction 

X_test
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
y_test
```

# 4. Evaluate the model

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

```python
# 5. improve model

# try different amount of n_estimators

np.random.seed(42)

for i in range (10,100,10):
    print(f"Trying model with {i} estimators.....")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train,y_train)
    print(f"Model Accuracy on test set: {clf.score(X_test,y_test) * 100:.2f}%")
```

```python
# 6 Save a model and load it 


import pickle

pickle.dump(clf,open("randm+forest_model_1.pkl","wb"))
```

```python
loaded_model = pickle.load(open("randm+forest_model_1.pkl","rb"))
loaded_model.score(X_test,y_test)
```
